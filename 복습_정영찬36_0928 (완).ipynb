```python
import seaborn as sns
tt = sns.load_dataset('titanic')
```

# Query

```python
# 컬럼의 정보 출력
tt.info()
```

```python
# 나이가 30세 이상인 데이터만 추출
tt.query('age >= 30')
```

```python
# 문자열을 추출하고싶을 경우, 바깥을 "로 두르고, 찾고싶은 문자열을 '로 두른다
# 여기서, class는 예약어이므로, ` (백틱)으로 감싼다.
# class가 first이면서 30세 이하 승객 추출
tt.query("`class` == 'First' & age<=30")
```

```python
# 20대 남성 이거나 / 여성 이면 추출
tt.query("age>=20 & age<=29 & sex=='male' | sex=='female'")
```

## 데이터프레임 재구조화
### 출력이 데이터프레임이면, 다시 데이터프레임으로 재사용할 수 있다.

```python
# 60세 이하 남성을 추출한 후, 그 값에서 class==First 값을 추출한다. 즉, 데이터프레임을 재사용한다.
tt.query("age<=60 & sex=='male'").query("`class`=='First'")
```

```python
# 30세 이하 남성 -> 생존자 -> 나이순 정렬 -> 최고령 5인 추출
tt.query("age<=50 & sex=='male'").query('survived==1').sort_values(by='age').iloc[82:]
```

```python
# 나이가 50이면 true, true인 값만 추출
tt[tt['age']==50]
```

```python
# 여성, 20세 이상이 true일 경우만 추출 후 3명 추출
tt[(tt['sex']=='female') & (tt['age']>=20)].iloc[:3]
```

```python
## 타이타닉 데이터의 남성과 여성 데이터를 추출하고 
## 남성도 여성도 pclass == 3 , age 20세 미만의 생존율
```

```python
# 남성 테이블
live_ratio_male = tt[(tt['sex']=='male') & (tt['pclass']==3) & (tt['age']<20)][['sex', 'survived']]
live_ratio_male
```

```python
# 여성 테이블
live_ratio_female = tt[(tt['sex']=='female') & (tt['pclass']==3) & (tt['age']<20)][['sex', 'survived']]
live_ratio_female
```

# 정렬
#### sort_values()를 이용한다.
#### 디폴트값은 오름차순이며, ascending=False를 통해 내림차순 정렬
#### NaN값은 마지막에 처리한다.

```python
# 오름차순 정렬
tt.sort_values('age')[['age','survived','who']]
```

```python
# 내림차순 정렬
tt.sort_values('age', ascending=False)[['age','survived','who']]
```

```python
# 두개 이상 정렬도 가능하다. 앞 컬럼이 동일값이면 뒷컬럼 기준 정렬
# 앞 컬럼은 오름차순, 뒷컬럼은 내림차순 정렬
tt.sort_values(['age','pclass'], ascending=[True, False])[['age','pclass','survived','who']]
```
