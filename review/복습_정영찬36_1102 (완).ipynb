# Merge

```python
test1 = pd.DataFrame({'학회원번호': [1,2,3,4,5],
                     '학회원분반':['python', 'sql', 'r', '데이터분석입문ml', '데이터분석입문통계']})
```

```python
test2 = pd.DataFrame({'학회원번호': [1,2,3,4,5],
                     '상벌점현황': [10, 20, 30, 40, 50]})
```

```python
display(test1, test2)
```

```python
# merge를 통한 병합
# 공통항목 즉, 학회원번호 기준으로 병합

pd.merge(test1, test2, how='inner')
```

```python
pd.merge(test2, test1)
```

```python
# 공통된 컬럼의 값이 다를 때

test_a=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['python','sql','r','데이터분석입문ml','데이터분석입문통계']})


test_b= pd.DataFrame({'학회원번호':[1,2,3,8,7],
             '상벌점현황':[10,20,30,40,50]})
```

```python
display(test_a, test_b)
```

```python
# inner: 공통된 부분만 포함(교집합)
# how 파라미터 미입력시 기본값은 inner
pd.merge(test_a, test_b, how='inner')
```

```python
# left: test_a 항목 모두 포함
pd.merge(test_a, test_b, how='left')
```

```python
# right: test_b 항목 모두 포함
pd.merge(test_a, test_b, how='right')
```

```python
# outer: 모든 부분 포함(합집합)
pd.merge(test_a, test_b, how='outer')
```

# Concat

```python
# 데이터프레임 여러개를 통째로 합치는 기능

# 기본값은 행 방향으로 결합
pd.concat([test1, test2])
```

```python
# 열 방향으로 결합
pd.concat([test1, test2], axis=1)
```

# Groupby

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
# pclass가 같은 값끼리 묶기
pcl_grp = df.groupby('pclass')
```

```python
# i: pclass의 값(1, 2, 3)
# j: 해당 pclass의 데이터프레임
for i, j in pcl_grp:
    print("pclass:", i)
    print(j)
```

```python
# pclass가 3인 그룹
pcl_grp.get_group(3)
```

```python
# 여러 열 같이 묶기
pcl_grp2 = df.groupby(['survived', 'sex'])
```

```python
# 출력
for i, j in pcl_grp2:
    print('그룹', i)
    print(j)
```

```python
# 사망한 남성의 평균 나이
pcl_grp2.get_group((0, 'male'))['age'].mean()
```

```python
# 등급과 성별로로 묶기
pcl_grp3 = df.groupby(['pclass', 'sex'])
```

```python
pcl_grp3[['age', 'fare']].mean()
```

```python
pcl_grp3[['age', 'fare']].agg(['mean', 'max', 'median'])
```

```python
def min_max(x):
    return x.max() - x.min()
```

```python
pcl_grp3[['age', 'fare']].agg(min_max)
```
