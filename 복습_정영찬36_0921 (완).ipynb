## 파생변수

```python
food = pd.DataFrame({'상품명': ['라면', '우유', '참치회', '통조림'],
                    '입고 수량': [100, 10, 5, 30],
                    '판매 수량': [20, 4, 5, 10]})
food
```

```python
# 입고 수량에서 판매 수량을 뺀 남은 수량 컬럼을 추가한다
food['남은 수량'] = food['입고 수량'] - food['판매 수량']
food
```

```python
# for문을 통한 전처리
# 남은 수량이 0이면 배열에 Y 삽입, 아니면 N 삽입
sold_out = []
for i in food['남은 수량']:
    if i==0:
        sold_out.append('Y')
    else:
        sold_out.append('N')
sold_out
```

```python
# where()문을 통한 데이터 서치
food['완판 여부'] = np.where(food['남은 수량']==0, 'Y', 'N')
food
```

```python
food['완판 여부'].value_counts()
```

## 행 기준 데이터 추출: loc, iloc

```python
tt
```

```python
# loc: 행과 라벨이름을 통한 데이터 선택
# 0~100번행 / 성별, 요금, 연령 열 추출
loc_tt = tt.loc[:100, ['sex', 'fare', 'age']]
loc_tt
```

```python
# iloc: 넘버링된 정수 인덱스를 통한 데이터 선택
# 0~99 행(총 100개) / 1번째, 3번째, 4번째 열 추출
iloc_tt = tt.iloc[:100, [0, 2, 4]]
iloc_tt
```

```python
tail_tt = tt.tail(100)
tail_tt
```

```python
len(tail_tt)
```

```python
# loc은 원본을 참조하기 때문에 tail_tt의 길이가 100에도 불구하고 오류가 발생하지 않는다
tail_tt.loc[794]
```

```python
# iloc은 현재 df를 참조하기 때문에 794번째 인덱스를 찾으려 하면 오류가 발생한다.
tail_tt.iloc(794)
```

```python
# iloc에서 794번 인덱스를 찾으려면 tail_tt에서의 인덱스. 즉 3번째 인덱스를 찾아야 한다.
tail_tt.iloc[3]
```
