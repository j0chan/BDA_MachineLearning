```python
import seaborn as sns
# tt변수에 타이타닉 데이터셋 저장
tt = sns.load_dataset('titanic')
# tt의 총 인원은 891명
tt
```

```python
# NaN값 제외
tt = tt.dropna()
```

```python
# 데이터프레임을 fare 기준 오름차순 정렬
fare_tt = tt.sort_values('fare')[['fare']]
```

```python
# 3등급으로 나누기 위해 1/3 지점, 2/3 지점 인덱스 추출
low_index = int(len(tt)/3)
high_index = int(low_index * 2)

print(low_index, high_index)
```

```python
# 인덱스를 기반으로 슬라이싱한 후, 각 배열에 삽입
# 등급을 low, mid, high로 나눈다.
low_fare = tt[:low_index] # 하위 297개 값 삽입
high_fare = tt[high_index:] # 상위 297개 값 삽입
mid_fare = tt[low_index:high_index] # 나머지 값 삽입
```

```python
# 전달값을 탐색하며 값이 1이면 count를 늘리는 함수.
# 이를 통해 위에서 추출한 값 중 생존한 인원을 체크한다.
def survived_num(a):
    count = 0
    for i in a:
        if i == 1:
            count += 1
    return count
```

```python
# low등급 남성 추출
low_fare_male = low_fare[low_fare['sex']=='male']
# low등급 여성 추출
low_fare_female = low_fare[low_fare['sex']=='female']
```

```python
# mid등급 남성 추출
mid_fare_male = mid_fare[mid_fare['sex']=='male']
# mid등급 여성 추출
mid_fare_female = mid_fare[mid_fare['sex']=='female']
```

```python
# high등급 남성 추출
high_fare_male = high_fare[high_fare['sex']=='male']
# high등급 여성 추출
high_fare_female = high_fare[high_fare['sex']=='female']
```

```python
# 결과 출력
print('----------남성----------')
print('low등급 남성의 수:', len(low_fare_male))
print('low등급 남성 생존수:', survived_num(low_fare_male['survived']))
print()
print('mid등급 남성의 수:', len(mid_fare_male))
print('mid등급 남성 생존수:', survived_num(mid_fare_male['survived']))
print()
print('high등급 남성의 수:', len(high_fare_male))
print('high등급 남성 생존수:', survived_num(high_fare_male['survived']))
print('\n----------여성----------')
print('low등급 여성의 수:', len(low_fare_female))
print('low등급 여성 생존수:', survived_num(low_fare_female['survived']))
print()
print('mid등급 여성의 수:', len(mid_fare_female))
print('mid등급 여성 생존수:', survived_num(mid_fare_female['survived']))
print()
print('high등급 여성의 수:', len(high_fare_female))
print('high등급 여성 생존수:', survived_num(high_fare_female['survived']))
```

### 각 등급의 남여 수의 합은 297로 동일하며, 세 등급을 모두 더하면 원본값 개수인 891개가 나온다.
